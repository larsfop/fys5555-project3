```python
analysis = '4lep'
```

```python
# Manipulate the simulated Monte Carlo data and choose which background and signal samples you want to use
mcfiles = initialize(opendatadir+"/"+analysis+"/MC")
datafiles = initialize(opendatadir+"/"+analysis+"/Data")
allfiles = z = {**mcfiles, **datafiles}
Backgrounds = getBkgCategories()

Backgrounds.remove('Zjets')
Backgrounds.remove('Wjets')
Backgrounds.remove('Wjetsincl')

signal = 'Higgs'
```

```python
files = []
for Category in Backgrounds:
    Type = mcfiles[Category]['type']
    for File in mcfiles[Category]['files']:
        files.append(File)
        
for File in datafiles['data']['files']:
    files.append(File)
```

```python
chain = R.TChain('mini') 

for File in files:
        chain.Add(File) 
```

```python
sel_ZZ = R.AnalysisSelector(chain, analysis)
```

```python
%%time

sel_ZZ.Process(signal)
```

```python
H_myy_d = R.TH1D()

H_myy_d.SetNameTitle('H_myy', 'Invariant mass')
H_myy_d.SetBins(24, 105, 160)

H_myy_d.Reset()

H_myy_d.Add(sel_yy.Selector().GetHistogram('myy', 'data'))
```

```python
Backgrounds.remove('Higgs')
```

```python
Backgrounds.remove('Diboson')

Backgrounds.append('Diboson')
Backgrounds.append('Higgs')
```

```python
H_mllll = {}

H_mllll['s+b'] = R.TH1D()
for bkg in Backgrounds:
    H_mllll[bkg] = R.TH1D()
    
    H_mllll[bkg].SetNameTitle('H_mllll', '4L invariant mass')
    H_mllll[bkg].SetBins(24, 80, 170)
    
    H_mllll[bkg].Reset()
    
    H_mllll[bkg].Add(sel_ZZ.Selector().GetHistogram('mllll', bkg)) # Aquire the specific histogram from the Selector pointer from the SelectorProxy class   
    
    H_mllll['s+b'] = R.TH1D()
    H_mllll['s+b'].SetNameTitle('H_mllll', '4L invariant mass')
    H_mllll['s+b'].SetBins(24, 80, 170)
    H_mllll['s+b'].Reset()
    H_mllll['s+b'].Add(sel_ZZ.Selector().GetHistogram('mllll', bkg))
    
```

```python
H_mllll_d = R.TH1D()

H_mllll_d.SetNameTitle('H_mllll', '4L invariant mass')
H_mllll_d.SetBins(24, 80, 170)

H_mllll_d.Reset()

H_mllll_d.Add(sel_ZZ.Selector().GetHistogram('mllll', 'data'))
```

```python
# turn fit into histogram with equal amount of bins to data histogram
bkg = sel_yy.Selector().GetFit('bkg')
bkg.SetNpx(24)
a = bkg.GetHistogram()
```

```python
bkgs = Backgrounds
bkgs.remove('Higgs')
```

```python
sig_eff_ZZ = sel_ZZ.Selector().GetSignalEfficiency()

sig_eff_yy = sel_yy.Selector().GetSignalEfficiency()
```

```python
Z = pyStats.countingExperimentWithBkgCorr(name='yy', intLum=10.06, intLumUnc=0.37)

# yy channel

# Signal region
n_obs = 0
b = 0
# Select bins corresponding to signal region
for i in range(8,11):
    n_obs += int(H_myy_d.GetBinContent(i))
    b += a.GetBinContent(i)
Z.addChannel('Signal region', bkg=b, bkgUncUncorr=0, bkgUncCorr=0, Nobs=n_obs, eff=sig_eff_yy)

# Control region
n_obs = 0
b = 0
for i in range(1,6):
    n_obs += int(H_myy_d.GetBinContent(i))
    b += a.GetBinContent(i)
Z.addChannel('Control region', bkg=b, bkgUncUncorr=0, bkgUncCorr=0, Nobs=n_obs, eff=0)

print(Z)

Z_yy = Z.getSignificance()
print('Significance =', Z_yy)

# ZZ channel

Z = pyStats.countingExperimentWithBkgCorr(name='ZZ', intLum=10.06, intLumUnc=0.37)

# Signal region
n_obs = 0
b = 0; b_err = 0
for i in range(11,14):
    n_obs += int(H_mllll_d.GetBinContent(i))
    for bkg in bkgs:
        b += H_mllll[bkg].GetBinContent(i)
        b_err += H_mllll[bkg].GetBinError(i)**2
Z.addChannel('Signal region', bkg=b, bkgUncUncorr=0, bkgUncCorr=b_err, Nobs=n_obs, eff=sig_eff_ZZ)

# Control region
n_obs = 0
b = 0; b_err = 0
for i in range(1,4):
    n_obs += int(H_mllll_d.GetBinContent(i))
    for bkg in bkgs:
        b += H_mllll[bkg].GetBinContent(i)
        b_err += H_mllll[bkg].GetBinError(i)**2
#Z.addChannel('Control region', bkg=b, bkgUncUncorr=0, bkgUncCorr=b_err, Nobs=n_obs, eff=0)
Z.setPriors(bkg='gauss');

print(Z)

Z_ZZ = Z.getSignificance()
print('Significance =', Z_ZZ)
```

```python
p_yy = 1 - sp.stats.norm.cdf(Z_yy)
p_ZZ = 1 - sp.stats.norm.cdf(Z_ZZ)

chi2 = -2*(np.log(p_yy) + np.log(p_ZZ))

p_com = 1 - sp.stats.chi2.cdf(chi2,4)

Z_com = sp.stats.norm.ppf(1-p_com)

print('Combined significance =', Z_com)
```

```python
R.gStyle.SetLegendBorderSize(0); ## Remove (default) border around legend 
leg = R.TLegend(0.65, 0.60, 0.9, 0.85); 

leg.Clear();
```

```python
C = R.TCanvas("c", "c", 700, 750)

w = R.RooWorkspace('w')

g_ZZ = w.factory('Gaussian::g_ZZ(x[122,128],mean[118,130],sigma1[0.1,50])')
g_yy = w.factory('Gaussian::g_yy(x,mean,sigma2[0.1,50])')
g = w.factory('Gaussian::g(x,mean,sigma3[0.1,50])')

x = w.var('x')
mean = w.var('mean')
```

```python
data_ZZ = R.RooDataHist("data", "data", x, H_mllll['s+b'])

nll_ZZ = g_ZZ.createNLL(data_ZZ)

minimizer = R.RooMinimizer(nll_ZZ)
minimizer.minimize('Minuit2')

m_ZZ = mean.getValV()
s_ZZ = mean.getError()

frame = mean.frame()

tmp = nll_ZZ.plotOn(frame, R.RooFit.ShiftToZero(), R.RooFit.LineColor(4))
p = tmp.getObject(0)

leg.AddEntry(p, 'H#rightarrow ZZ*#rightarrow 4l', 'l')

frame.Draw()
```

```python
data_yy = R.RooDataHist("data", "data", x, H_myy_d)

nll_yy = g_yy.createNLL(data_yy)

minimizer = R.RooMinimizer(nll_yy)
minimizer.minimize('Minuit2')

m_yy = mean.getValV()
s_yy = mean.getError()

tmp = nll_yy.plotOn(frame, R.RooFit.ShiftToZero(), R.RooFit.LineColor(2))
p = tmp.getObject(1)

leg.AddEntry(p, 'H#rightarrow #gamma#gamma', 'l')

frame.Draw()
```

```python
data = R.RooDataHist('data', 'data', x, R.RooFit.Import(H_myy_d), R.RooFit.Import(H_mllll['s+b']))

nll = g.createNLL(data)

minimizer = R.RooMinimizer(nll)
minimizer.minimize('Minuit2')

m_com = mean.getValV()
s_com = mean.getError()

tmp = nll.plotOn(frame, R.RooFit.ShiftToZero(), R.RooFit.LineColor(1))
p = tmp.getObject(2)

leg.AddEntry(p, 'Combined', 'l')

frame.Draw()
```
